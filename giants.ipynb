```python
import pandas as pd
import matplotlib.pyplot as plt
```

To get the stats for only the Giants hitters, use the Team column to select "SFG" as the team. Keep only hitters who have more than 100 projected at bats to try limit the analysis to only statisically significant players. Remove columns that are not of interest and players who did not actually play for the team in 2021.

```python
projected_2021_batters = pd.read_csv('Fangraphs Leaderboard Batters.csv')

projected_giants_2021_batters = projected_2021_batters.loc[projected_2021_batters['Team'] == 'SFG']
projected_giants_2021_batters = projected_giants_2021_batters.loc[projected_2021_batters['AB'] > 100] 
projected_giants_2021_batters = projected_giants_2021_batters.sort_values(by=['Name'], ascending = True)
projected_giants_2021_batters = projected_giants_2021_batters.reset_index(drop=True)
projected_giants_2021_batters = projected_giants_2021_batters.drop(['ADP',
                                                                    'playerid',
                                                                    '-1.3',
                                                                    '-1.2',
                                                                    '-1.1',
                                                                    '-1',
                                                                    'Def',
                                                                    'Off',
                                                                    'PA',
                                                                    'HBP',
                                                                    'WAR',
                                                                    'Fld',
                                                                    'BsR'], axis=1)
projected_giants_2021_batters = projected_giants_2021_batters.drop([9])
projected_giants_2021_batters = projected_giants_2021_batters.reset_index(drop=True)
projected_giants_2021_batters
```

Do the same thing as above with Giants pitchers, except keep only pitchers with more than 40 projected innings pitched.

```python
projected_2021_pitchers = pd.read_csv('Fangraphs Leaderboard Pitchers.csv')

projected_giants_2021_pitchers = projected_2021_pitchers.loc[projected_2021_pitchers['Team'] == 'SFG']
projected_giants_2021_pitchers = projected_giants_2021_pitchers.loc[projected_2021_pitchers['IP'] > 40]
projected_giants_2021_pitchers = projected_giants_2021_pitchers.sort_values(by=['Name'], ascending = True)
projected_giants_2021_pitchers = projected_giants_2021_pitchers.reset_index(drop=True)
projected_giants_2021_pitchers = projected_giants_2021_pitchers.drop(['ADP',
                                                                    'playerid',
                                                                    'RA9-WAR',
                                                                    'WAR'], axis=1)
projected_giants_2021_pitchers = projected_giants_2021_pitchers.drop([6,9,10,11,12,13])
projected_giants_2021_pitchers = projected_giants_2021_pitchers.reset_index(drop=True)
projected_giants_2021_pitchers
```

Get the actual stats of the Giants batters as above, but keep players with over 80 at bats since the players have played less games than projected so far.

```python
actual_2021_batters = pd.read_csv('mlb-player-stats-Batters.csv')

actual_giants_2021_batters = actual_2021_batters.loc[actual_2021_batters['Team'] == 'SF'] 
actual_giants_2021_batters = actual_giants_2021_batters.loc[actual_2021_batters['AB'] > 80]
actual_giants_2021_batters = actual_giants_2021_batters.sort_values(by=['Player'], ascending = True)
actual_giants_2021_batters = actual_giants_2021_batters.reset_index(drop=True)
actual_giants_2021_batters = actual_giants_2021_batters.drop(['SH',
                                                              'SF',
                                                              'HBP'], axis=1)
actual_giants_2021_batters = actual_giants_2021_batters.drop([10,14,15])
actual_giants_2021_batters = actual_giants_2021_batters.reset_index(drop=True)
actual_giants_2021_batters
```

Do the same as above with the Giants pitchers. Innings pitched can be the same as the projected stats.

```python
actual_2021_pitchers = pd.read_csv('mlb-player-stats-Pitchers.csv')

actual_giants_2021_pitchers = actual_2021_pitchers.loc[actual_2021_pitchers['Team'] == 'SF']
actual_giants_2021_pitchers = actual_giants_2021_pitchers.loc[actual_2021_pitchers['IP'] > 40]
actual_giants_2021_pitchers = actual_giants_2021_pitchers.sort_values(by=['Player'], ascending = True)
actual_giants_2021_pitchers = actual_giants_2021_pitchers.reset_index(drop=True)
actual_giants_2021_pitchers = actual_giants_2021_pitchers.drop([9])
actual_giants_2021_pitchers = actual_giants_2021_pitchers.reset_index(drop=True)
actual_giants_2021_pitchers
```

OPS could be a good indicator of how well the Giants batters are playing. We can plot the projected OPS and the actual OPS for each player that appears in both the projected stats and the actual stats dataframes.

```python
x = actual_giants_2021_batters['Player']
y = projected_giants_2021_batters['OPS']
y2 = actual_giants_2021_batters['OPS']

df = pd.DataFrame(
    {'Player': x,
     'Projected OPS': y,
     'Actual OPS': y2
    })

df.plot(x='Player', y=['Projected OPS', 'Actual OPS'], kind='bar', ylabel='OPS', color=['C1', 'k'])

# N=9
# ind = np.arange(N)
# width = 0.35

# fig = plt.figure()
# ax = fig.add_axes([0,0,1,1])
# ax.bar(ind, y, width)
# ax.bar(ind+width, y2, width)
# plt.xticks(ind + width / 2, x)
# ax.legend(labels=['Projected Stats', 'Actual Stats'])
# plt.xlabel('Player Age')
# plt.ylabel('OPS')
# plt.setp(ax.get_xticklabels(), rotation = 90)
# plt.show()

# plt.plot(x, y, '.-', label='Projected OPS')
# plt.plot(x, y2, '.-', label='Actual OPS')

# plt.xlabel('Player')
# plt.ylabel('OPS')
# plt.legend()
# ax = plt.gca()
# plt.setp(ax.get_xticklabels(), rotation = 90)
# plt.show()
```

Based on domain knowledge, the players who are outperforming their projected OPS are the more veteran players. The source data has an age column, so the age of the players can be plotted against OPS. Based on the plot, we can see that players older than 32 are outplaying their projections moreso that players younger than 32.

```python
actual_giants_2021_batters = actual_giants_2021_batters.sort_values(by=['Age'], ascending = True)
actual_giants_2021_batters_age = actual_giants_2021_batters['Age']
projected_giants_2021_batters['Age'] = actual_giants_2021_batters_age
projected_giants_2021_batters = projected_giants_2021_batters.sort_values(by=['Age'], ascending = True)

x = actual_giants_2021_batters['Age']
y = projected_giants_2021_batters['OPS']
y2 = actual_giants_2021_batters['OPS']

# plt.plot(x, y, '.-', label='Projected OPS')
# plt.plot(x, y2, '.-', label='Actual OPS')

# plt.xlabel('Player Age')
# plt.ylabel('OPS')
# plt.legend()
# ax = plt.gca()
# plt.setp(ax.get_xticklabels(), rotation = 90)
# plt.show()

df = pd.DataFrame(
    {'Player': x,
     'Projected OPS': y,
     'Actual OPS': y2
    })

df.plot(x='Player', y=['Projected OPS', 'Actual OPS'], kind='bar', xlabel='Player Age', ylabel='OPS', rot=0, color=['C1', 'k'])
```
